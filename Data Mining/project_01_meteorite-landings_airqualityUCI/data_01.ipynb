## Cách làm sạch dữ liệu

```python
# Đổi tên các cột trong DataFrame
df.columns 
```

```python
# Kiểm tra dữ liệu với giá trị null
# trong đó True sẽ là đại diện cho các ô có gía trị null(rỗng)
# Trong khi False được sử dụng để đại diện cho các ô có giá trị khác null
df.isnull()
```

```python
#Kiểm tra dữ với giá trị khác null
df.notnull()
```

```python
# Cách bỏ toàn bộ các ô có dữ liệu nullk
df.dropna()
```

```python
# Cách bỏ toàn cột các ô có dữ liệu null
df.dropna(axis=1)
```

```python
# Cách bỏ các dòng có nhiều hơn n giá trị null
df.dropna(axis=1, thresh=3)
```

```python
# Cách thay toàn bộ giá trị null bằng giá trị index_x
# Trong Vd này thay thế toàn bộ giá trị null  bằng giá trị trung bình của từng cột
df.fillna(df.mean())
```

```python
# Cách thay toàn bộ giá trị null kahcs bằng giá trị khác
# ở đây thay tất cả các giá trị bằng 1 thay thế bằng 'one'
df.replace(1, 'one') 
```

```python
#Cách thay nhiều giá trị cùng một lúc
df.replace( [1,3],['one', 'three'])
```

```python
# Thay đổi tên cột
df.rename(columns={'Time': 'Thoi Gian'})
```

```python
#Cách đổi index trong DataFrame
df.set_index('Date', inplace=True)
```

## Cách lọc, sắp xếp, nhóm dữ liệu trong pandas DataFrame

```python
#Lọc dữ liệu theo điều kiện
# Lọc ra các dòng thoả mãn điều kiện col lớn hơn 100
df[df['NMHC(GT)'] > 100]
```

```python
# Lọc ra các dòng thoả mãn điều kiện Có giá tị col trong khoảng 100 đến 200
df[(df['NMHC(GT)'] > 100) & (df['NMHC(GT)'] < 200)]
```

```python
#  sắp xếp dữ liệu trong cột col1 theo chiều thuận(ascending)
# Theo thứ tự tăng dần
df.sort_values('NMHC(GT)')
```

```python
# Theo thứ tự giảm dần
df.sort_values('NMHC(GT)', ascending=False)  
```

```python
# sắp xếp col1 và col2 theo chiều thuận và theo chiều nghịch
df.sort_values(['NMHC(GT)','PT08.S1(CO)'], ascending=[True, False])
```

```python
#  Tính mean của tất cả các cột 
df.mean()
```

```python
# Tính max cho tất cả các dòng
df.max()
```
