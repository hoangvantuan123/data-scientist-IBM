#Vẽ biểu đồ Histogram cho cột 'CO(GT)'

* ***bins=20*** : Tham số này xác định số lượng khoảng (bins) trong histogram. Trong trường hợp này ta chia dữ liệu thành 20 khoảng

## Biểu đồ tương quan
 * ***correlation = data_dropna.corr()***: Đây là phép tính ma trận tương quan giữa các cặp biến trong DataFrame data. Kết quả là một DataFrame mới chứa các giá trị tương quan giữa các cặp biến.
 *  ***plt.figure(figsize=(10, 10))***: Đây là lệnh tạo một hình ảnh (figure) có kích thước 10x10 inch để chứa biểu đồ.
 * ***plt.imshow(correlation, cmap='coolwarm', interpolation='none')***: Đây là lệnh vẽ biểu đồ tương quan bằng cách hiển thị ma trận tương quan (correlation). Tham số cmap='coolwarm' xác định bảng màu được sử dụng (coolwarm là một bảng màu phổ biến). Tham số interpolation='none' cho biết không có sự nội suy giữa các ô.
 * ***plt.colorbar()***: Đây là lệnh hiển thị thanh màu cho biểu đồ tương quan, giúp giải thích các giá trị màu tương ứng với mức độ tương quan.
 * ***plt.xticks(range(len(correlation)), correlation.columns, rotation=90)***: Đây là lệnh đặt nhãn cho trục x của biểu đồ tương quan. Các nhãn được lấy từ tên cột trong ma trận tương quan (correlation.columns) và được hiển thị dọc theo trục x với góc quay 90 độ.
 * ***plt.yticks(range(len(correlation)), correlation.columns)***: Đây là lệnh đặt nhãn cho trục y của biểu đồ tương quan. Các nhãn được lấy từ tên cột trong ma trận tương quan (correlation.columns) và được hiển thị dọc theo trục y.




````python
correlation = data_dropna.corr()
plt.figure(figsize=(10, 10))
plt.imshow(correlation, cmap='coolwarm', interpolation= 'none')
plt.colorbar()
plt.xticks(range(len(correlation)), correlation.columns, rotation = 90)
plt.yticks(range(len(correlation)), correlation.columns )

plt.title('Correlation Matrix')
plt.show()
````

````python
description = data_dropna.describe()


correlation = description.corr()
plt.figure(figsize=(10, 10))
plt.imshow(correlation, cmap='coolwarm', interpolation= 'none')
plt.colorbar()
plt.xticks(range(len(correlation)), correlation.columns, rotation = 90)
plt.yticks(range(len(correlation)), correlation.columns )

plt.title('Correlation Matrix')
plt.show()
````

* ***plt.title('CO(GT) Counts'), plt.xlabel('CO(GT)'), plt.ylabel('Count')***: Đoạn mã này đặt tiêu đề cho biểu đồ (title), nhãn cho trục x (xlabel) và nhãn cho trục y (ylabel) để giải thích ý nghĩa của biểu đồ.

plt.title('Category Distribution'), plt.axis('equal'): Đoạn mã này đặt tiêu đề cho biểu đồ (title) và đảm bảo biểu đồ pie chart có hình dạng hình tròn (axis('equal')).

### Cách này chưa tối ưu.  vì nếu thế làm cho tất cả các giá trị bằng nhau

````python
# Tinh tong tat ca gia tri cac cot
# Su dung ham lambda de tinh tong  gai tri duy nhat trong 1 cot
value_counts = data_dropna.apply(lambda x: x.value_counts().sum())  

plt.figure(figsize=(10, 6))
plt.pie(value_counts.values, labels=value_counts.index, autopct='%1.1f%%', startangle= 90 )
plt.title("Text")
plt.axis('equal')
plt.tight_layout()
plt.show()
````
